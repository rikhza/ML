# Simple linear regression 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
# Read CSV using Pandas
Data=pd.read_csv('Startups.csv')
Data
```

```python
# Drop Zero Values
Data = Data[Data!=0].dropna()
Data
```

```python
# Encode State Columns
New_data = pd.get_dummies(Data, columns=['State'])
New_data
```

```python
New_data.describe()
```

```python
# Split Value to X and Y
X = New_data.drop(columns=['Profit'])
Y = New_data['Profit']
```

```python
# Use train_test_split function to automatically split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
# Call Linear Regression Function
Regression = LinearRegression()
```

```python
# Start Calculating
Regression.fit(x_train, y_train)
```

```python
# Predict and Show Values between Predicted and Actual Value
y_predict = Regression.predict(x_test)
print(pd.array(y_predict))
print(pd.array(y_test))
```

```python
mse = mean_squared_error(y_test, y_predict)
r2 = r2_score(y_test, y_predict)
print(mse, r2)
```
